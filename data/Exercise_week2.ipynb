```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sklearn.linear_model as lm

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

dataset1 = pd.read_csv("ds1.csv")
dataset2 = pd.read_csv("ds2.csv")
dataset3 = pd.read_csv("ds3.csv")
dataset4 = pd.read_csv("ds4.csv")
dataset5 = pd.read_csv("ds5.csv")

#Using Linear Regression model

#fitting the entire data
lnr1 = lm.LinearRegression()
lnr1.fit(dataset1.x.reshape(-1,1), dataset1.y)

lnr2 = lm.LinearRegression()
lnr2.fit(dataset2.x.reshape(-1,1), dataset2.y)

lnr3 = lm.LinearRegression()
lnr3.fit(dataset3.x.reshape(-1,1), dataset3.y)

lnr4 = lm.LinearRegression()
lnr4.fit(dataset4.x.reshape(-1,1), dataset4.y)

lnr5 = lm.LinearRegression()
lnr5.fit(dataset5.x.reshape(-1,1), dataset5.y)


print("dataset1 intercept" + str(lnr1.intercept_))
print("dataset1 coeff" + str(lnr1.coef_))

print("dataset2 intercept" + str(lnr2.intercept_))
print("dataset2 coeff" + str(lnr2.coef_))

print("dataset3 intercept" + str(lnr3.intercept_))
print("dataset3 coeff" + str(lnr3.coef_))

print(lnr4.intercept_)
print(lnr4.coef_)

print(lnr5.intercept_)
print(lnr5.coef_)

#plotting the entire data

plt.figure()
sns.regplot(dataset1.x, dataset1.y, order=1, ci=None, scatter_kws={'color':'red', 's':9}, data=dataset1)

plt.figure()
sns.regplot(dataset2.x, dataset2.y, order=1, ci=None, scatter_kws={'color':'red', 's':9}, data=dataset2)

plt.figure()
sns.regplot(dataset3.x, dataset3.y, order=1, ci=None, scatter_kws={'color':'red', 's':9}, data= dataset3)

plt.figure()
sns.regplot(dataset4.x, dataset4.y, order=1, ci=None, scatter_kws={'color':'red', 's':9}, data= dataset4)

plt.figure()
sns.regplot(dataset5.x, dataset5.y, order=1, ci=None, scatter_kws={'color':'red', 's':9}, data= dataset5)

#separating data to test and train

X_train, X_test, y_train, y_test = train_test_split(dataset1.x, dataset1.y, test_size=.2, random_state=0)

#Fitting the training data
lnr = lm.LinearRegression()
lnr.fit(X_train.reshape(-1,1),y_train)
plt.figure()
sns.regplot(X_train,y_train, order=1, ci=None, scatter_kws={'color':'red', 's':9})

#Assessing the accuracy of the model
print("RMSE: ", mean_absolute_error(y_test, lnr.predict(X_test.reshape(-1,1)))**(.05))
print("R^2: ", r2_score(y_test, lnr.predict(X_test.reshape(-1,1))))


#Plotting the results
plt.figure()
sns.regplot(X_test, y_test, order=1, ci=None, scatter_kws={'color':'red', 's':9})
plt.figure()
sns.regplot(X_test, lnr.predict(X_test.reshape(-1,1)),order=1, ci=None, scatter_kws={'color':'red', 's':9})


#Using KNN
from sklearn.neighbors import KNeighborsRegressor

kn_model = KNeighborsRegressor()
kn_model.fit(X_train.reshape(-1,1), y_train)

y_predicted = kn_model.predict(X_test.reshape(-1,1))

print("RMSE: ", mean_absolute_error(y_test, y_predicted)**(.05))
print("R^2: ", r2_score(y_test, y_predicted))

plt.figure()
sns.regplot(X_test, kn_model.predict(X_test.reshape(-1,1)),order=1, ci=None, scatter_kws={'color':'red', 's':9})


kn_model1 = KNeighborsRegressor()
kn_model1.fit(dataset4.x.reshape(-1,1), dataset4.y)

y_predicted = kn_model1.predict(dataset4.x.reshape(-1,1))

#print("RMSE: ", mean_absolute_error(y_test, y_predicted)**(.05))
#print("R^2: ", r2_score(y_test, y_predicted))

plt.figure()
sns.regplot(dataset4.x, kn_model.predict(dataset4.x.reshape(-1,1)),order=1, ci=None, scatter_kws={'color':'red', 's':9})


```
